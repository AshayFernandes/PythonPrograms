```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

black=pd.read_csv('blackfri.csv')
black.head()
```

```python
black_copy=black.drop(['User_ID','Product_ID','Stay_In_Current_City_Years'],axis=1,inplace=False)
black_copy.head()
```

```python
black['City_Category']=black['City_Category'].fillna('Not Applicable')
black.head()
```

```python
black['City_Category']=black['City_Category'].map({
            'A':'Metro Cities',
             'B':'Small Town',
              'C':'Villages' 
})
black.head()
```

```python
black.rename(columns={'Baseball_Caps':'Ball_Caps'},inplace=True)
black.head()             
```

```python
black['Marital_Status']=black['Marital_Status'].map({
                0:'un-married',1:'married'
})
black.head()
```

```python
import math
def tall(row):
    if (math.isnan(row.Ball_Caps) or math.isnan(row.Wine_Tumblers)):
        return 'no'
    else:
        return 'yes'
black.loc[:,'result']=black.apply(tall,axis=1)    
ax=sns.countplot(x='result',hue='Gender',data=black)
plt.show()
```
